```
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

**You can set the background of the graph to Darkgrid, White and Dark.
Uncomment each line of code to see the outcome.**

```
sns.set(style = 'darkgrid')
#sns.set(style = 'white')
#sns.set(style = 'dark')
```

```
tips = sns.load_dataset('tips')
tips.tail()        # print the last first data entry 
```

```
sns.relplot(x = 'total_bill', y = 'tip', data = tips)
```

```
#print the aggregate of smokers in the dataset
tips['smoker'].value_counts()
```

```
#hue is use to diiferent datapoint with colour
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker')
```

```
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker', style = 'time')
```

```
sns.relplot(x = 'total_bill', y = 'tip', style = 'size', data = tips)
```

```
#pallete is use to tune the colour of the graph
sns.relplot(x = 'total_bill', y = 'tip', hue = 'size', data = tips, palette = 'ch:r=-0.7, l= 0.95')
```

```
sns.relplot(x = 'total_bill', y = 'tip', data = tips, size = 'size')
```

```
#sizes is use to regualte the size of the legend
sns.relplot(x = 'total_bill', y = 'tip', data = tips, size = 'size' ,sizes = (15, 150))
```

```
sns.relplot(x = 'total_bill', y = 'tip', kind = 'line', data = tips, sort = True)
```

```
sns.relplot(x = 'total_bill', y = 'tip', kind = 'line', data = tips, sort = False)
```

```
dots = sns.load_dataset('dots').query("align == 'dots'")
dots.head()
```

```
sns.relplot(x = 'time', y = 'firing_rate', data = dots, kind = 'line', hue = 'coherence', style = 'choice')
```

```
sns.relplot(x = 'time', y = 'firing_rate', hue = 'coherence', size = 'choice', style = 'choice', kind = 'line', data = dots)
```
